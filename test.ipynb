## Read ERA5 file

```python
ds = xr.open_dataset('ERA5.nc')
ds # ds as dataset
```

```python
# Let's try to get the temperature in °C
da = ds.t2m - 273.15
da
```

Oups the kernel as shut down! This is because we have a very limited available amount of RAM (2Go here on binder). So we are going to have to trick to make this computation... Hopefully `xarray` comes with `dask` that allows easy parallel computation. Here we are not going to use the parallelization, but we are going to take advantage of Dask for splitting our data into multiple chunks to reduce to RAM usage!

See more: http://xarray.pydata.org/en/stable/user-guide/dask.html

```python
ds = xr.open_dataset('ERA5.nc', chunks={"longitude": 360, "latitude": 360})
ds # ds as dataset
```

```python
da = ds.t2m
da
```

You can see that our dataset have been splited into 12 chunks of about 261 MB what is better than the 2 Go full array! We are first going to compute the climatology and then make opration!

```python
clim = da.mean('time').load() # Load allow to load the result into memory
clim.plot()
```

```python
cmap='ColdHot'
levels=plot.arange(-30,30,5)
extend='both'

fig, axs = plot.subplots(nrows=1, ncols=1, proj='cyl', axwidth=5)

axs[0].contourf(clim - 273.15, colorbar='r', cmap=cmap, levels=levels, extend=extend, colorbar_kw={'label': 'Near-surface air temperature [°C]'})

axs.format(
    labels=True, 
#     lonlines=10, latlines=5, 
#     latlim=(latlim.start, latlim.stop), lonlim=(lonlim.start, lonlim.stop),
    coast=True, 
#     ocean=True, oceancolor='gray3', 
    borders=True,
    suptitle='ERA5 near-surface air temperature climatology (1979-2020)'
)
```
